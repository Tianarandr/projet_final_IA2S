```
import os.path
# select dataset GaCo / JuCo / SiCo
for subject in range(1,50):
  data='GaCo'
  if subject<10:
    file_name= './Segmented GaitCycles (multidim)/'+data+ str(subject).zfill(2)+ '_01_new.mat'
  else:
     file_name= './Segmented GaitCycles (multidim)/'+ data + str(subject)+ '_01_new.mat'

  if  os.path.exists(file_name):
  
# reading data from folder
    ndata = {}
    mat = loadmat(file_name)  # load mat-fil
    
    file_name = file_name.replace('./Segmented GaitCycles (multidim)/' ,'')
    df = pd.DataFrame(mat[ file_name[:-4] ],
                      columns=['Time', 'LxVGRF1', 'LxVGRF2', 'LxVGRF3', 'LxVGRF4', 'LxVGRF5', 'LxVGRF6', 'LxVGRF7', 'LxVGRF8', 'RxVGRF1', 'RxVGRF2', 'RxVGRF3', 'RxVGRF4', 'RxVGRF5', 'RxVGRF6', 'RxVGRF7', 'RxVGRF8', 'Lx Total Force', 'Rx Total Force'])

# Total force under the left foot
    left = df['Lx Total Force'].values
    left [left < 20] = 0
 
# Total force under the left foot
    right = df['Rx Total Force'].values
    right [right < 20] = 0

#split the data into swing and stance phases by grouping the forces equal to zero and the forces not equal to zero 
    l = left
    l = [list(g) for k, g in groupby(l, key=lambda x:x!=0)]
    l = np.array(l)

  
    r = right
    r = [list(g) for k, g in groupby(r, key=lambda x:x!=0)]
    r = np.array(r)

 
    lco.append(left)
    rco.append(right)

#calculation of the stride cycle left foot 
    l_result = []
    if all([v == 0 for v in l[0]]):
      l1 = l[list(range(1, len(l), 2))]
      l2 = l[list(range(2, len(l), 2))]
      for element1, element2 in zip(l1, l2):
        l_result.append(element1+element2)
    else:
      l1 = l[list(range(0, len(l), 2))]
      l2 = l[list(range(1, len(l), 2))]
      for element1, element2 in zip(l1, l2):
        l_result.append(element1 + element2)

#calculation of the stride cycle right foot
    r_result = []
    if all([v == 0 for v in r[0]]):
      r1 = r[list(range(1, len(r), 2))]
      r2 = r[list(range(2, len(r), 2))]
      for element1, element2 in zip(r1, r2):
        r_result.append(element1+element2)
    else:
      r1 = r[list(range(0, len(r), 2))]
      r2 = r[list(range(1, len(r), 2))]
      for element1, element2 in zip(r1, r2):
        r_result.append(element1 + element2)


#mean/std duration of stride time of the left foot 
    l_stride = []
    for element in l_result[1:]:
      l_stride.append(len(element))

    l_mean_stride=np.mean(l_stride)
    l_std_stride=np.std(l_stride)
    print('mean in duration of the stride time of the left foot', l_mean_stride,'\nstandad deviation of the stride time of the left foot',l_std_stride)
    print('\n')

#mean/std duration of stride time of the right foot 
    r_stride = []
    for element in r_result[1:]:
      r_stride.append(len(element))

    r_mean_stride=np.mean(r_stride)
    r_std_stride=np.std(r_stride)
    print('mean in duration of the stride time of the right foot', r_mean_stride,'\nstandard deviation of the stride time of the right foot',r_std_stride)


#mean/std in duration of the swing time of the left foot
    l_swing = []
    for element in l2[1:]:
      l_swing.append(len(element))
    l_mean_swing=np.mean(l_swing)
    l_std_swing=np.std(l_swing)
    print('mean in duration of the swing time of the left foot', l_mean_swing,'\nstandard deviation of the swing time of the left foot',l_std_swing)
    print('\n')

#mean/std in duration of the swing time of the left foot
    r_swing = []
    for element in r2[1:]:
      r_swing.append(len(element))
    r_mean_swing=np.mean(r_swing)
    r_std_swing=np.std(r_swing)
    print('mean in duration of the swing time of the right foot', r_mean_swing,'\nstandard deviation of the swing time of the right foot',r_std_swing)

#mean/std in duration of the stance time
    s= left*right
    s = [list(g) for k, g in groupby(s, key=lambda x:x!=0) if k]
    s = np.array(s)

    stance = []
    for element in s[1:]:
      stance.append(len(element))
    mean_stance=np.mean(stance)
    std_stance=np.std(stance)
    print('mean in duration of the stance time', mean_stance, '\nstandard deviation of the stancce time',std_stance)


    l_mean_stride=l_mean_stride/100
    l_std_stride=l_std_stride/100
    l_mean_swing=l_mean_swing/100
    l_std_swing=l_std_swing/100

    r_mean_stride=r_mean_stride/100
    r_std_stride=r_std_stride/100
    r_mean_swing=r_mean_swing/100
    r_std_swing=r_std_swing/100


    mean_stance=mean_stance/100
    std_stance=std_stance/100

#Coefficient of variation in percentage/duration of the swing time of the left foot
#Coefficient of variation in duration of the stride time of the left foot

    mean_per_swing_left=(l_mean_swing/l_mean_stride)*100
    std_per_swing_left=(l_std_swing/l_std_stride)*100

    var_per_swing_left= (std_per_swing_left/mean_per_swing_left)*100
    var_dur_swing_left=  (l_std_swing/l_mean_swing)*100
    var_dur_stride_left= (l_std_stride/l_mean_stride)*100
    print('Coefficient of variation in percentage of the swing time of the left foot', var_per_swing_left ,'\nCoefficient of variation in duration of the swing time of the left foot',var_dur_swing_left, '\nCoefficient of variation in duration of the stride time of the left foot',var_dur_stride_left)


    print('\n')

#Coefficient of variation in percentage/duration of the swing time of the right foot
#Coefficient of variation in duration of the stride time of the right foot
    mean_per_swing_right=(r_mean_swing/r_mean_stride)*100
    std_per_swing_right=(r_std_swing/r_std_stride)*100

    var_per_swing_right= (std_per_swing_right/mean_per_swing_right)*100 
    var_dur_swing_right=  (r_std_swing/r_mean_swing)*100
    var_dur_stride_right= (r_std_stride/r_mean_stride)*100
    print('Coefficient of variation in percentage of the swing time of the right foot', var_per_swing_right ,'\nCoefficient of variation in duration of the swing time of the right foot',var_dur_swing_right, '\nCoefficient of variation in duration of the stride time of the right foot',var_dur_stride_right)

#mean in percentage of the double stance time 
    mean_per_double_stance=(mean_stance/l_mean_stride)*100
    print('\mean in percentage of the double stance time ',mean_per_double_stance)


#long swing time and short swing time
    if (l_mean_swing > r_mean_swing) :
      short_swing=r_mean_swing
      long_swing=l_mean_swing
      var_short_swing= (r_std_swing/r_mean_swing)*100
      var_long_swing= (l_std_swing/l_mean_swing)*100
    else :
      short_swing=l_mean_swing
      long_swing=r_mean_swing
      var_short_swing= (l_std_swing/l_mean_swing)*100
      var_long_swing= (r_std_swing/r_mean_swing)*100

#Coefficient of variation of short/long swing time
    print('Coefficient of variation of short swing time ',var_short_swing,'\nCoefficient of variation of long swing time',var_long_swing)
    mean_gait_asymmetry= abs(math.log(short_swing/long_swing)) *100
    var_gait_asymmetry= abs(math.log(var_short_swing/var_long_swing)) *100

#Coefficient variation of the gait asymmetry
    print('\nCoefficient variation of the gait asymmetry', var_gait_asymmetry)

#19 features + label
    fea=[var_per_swing_left,var_dur_swing_left,var_dur_stride_left,var_per_swing_right,var_dur_swing_right,var_dur_stride_right,var_short_swing,var_long_swing,var_gait_asymmetry,mean_per_swing_left
                    ,l_mean_swing,l_mean_stride, mean_per_swing_right,r_mean_swing,r_mean_stride,mean_per_double_stance,short_swing,long_swing,mean_gait_asymmetry,0]
  
    co.append(fea)
    

  else:
    continue
```

```
from keras.models import Sequential                           # Model type to be used
from keras.layers import Input, Activation, Dense , Dropout
from keras.optimizers import SGD
from keras.layers.normalization import BatchNormalization
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

# Define the K-fold Cross Validator
kfold = KFold(n_splits=4, shuffle=True)

# K-fold Cross Validation model evaluation
fold_no = 1
acc_per_fold=[]
loss_per_fold=[]


model = Sequential()
model.add(Dense(32, input_dim=12, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))

# compile the keras model
model.compile(loss='binary_crossentropy', optimizer=SGD(), metrics=['accuracy'])
model.save_weights('./weights.h5')


for train, test in kfold.split(X, Y):

 

  # Generate a print
  print('------------------------------------------------------------------------')
  print(f'Training for fold {fold_no} ...')

  # Fit data to model
  model.load_weights('./weights.h5') 
  es = EarlyStopping(monitor='val_loss', mode='min', patience = 5)
  mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', verbose=1, save_best_only=True)
  history = model.fit(X[train], Y[train], batch_size=32, epochs=1000, verbose=1, callbacks=[es, mc])

  # Generate generalization metrics
  scores = model.evaluate(X[test], Y[test], verbose=0)
  print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
  acc_per_fold.append(scores[1] * 100)
  loss_per_fold.append(scores[0])

  # Increase fold number
  fold_no = fold_no + 1

```
